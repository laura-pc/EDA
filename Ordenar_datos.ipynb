Una vez tenemos los datos, vamos acrear tablas, ordenarlos y prepararlos para el análisis

```python
import pandas as pd
import numpy as np
```

### Creamos data frame de airbnb y ordenamos datos

```python
a_marzo = pd.read_csv("airbnb_marzo_2023", index_col= 0)
a_abril = pd.read_csv("airbnb_abril_2023", index_col= 0)
a_mayo = pd.read_csv("airbnb_mayo_2023", index_col= 0)
a_junio = pd.read_csv("airbnb_junio_2023", index_col= 0)
a_julio = pd.read_csv("airbnb_julio_2023", index_col= 0)
a_agosto = pd.read_csv("airbnb_agosto_2023", index_col= 0)
a_sept = pd.read_csv("airbnb_sept_2023", index_col= 0)
a_oct = pd.read_csv("airbnb_oct_2023", index_col= 0)
a_nov = pd.read_csv("airbnb_nov_2023", index_col= 0)
a_dic = pd.read_csv("airbnb_dic_2023", index_col= 0)
a_enero = pd.read_csv("airbnb_enero_2024", index_col= 0)
a_feb = pd.read_csv("airbnb_feb_2024", index_col= 0)
```

Como la forma de ordenar los datos que he utilizado para los hoteles no funcionaba con estos, ya que salían los numeros raros y con muchos decimales usé estos pasos para colocar los dataframes. 

```python
#a_feb = list(a_feb.stack())
#a_feb.sort()
#a_feb = pd.DataFrame(a_feb, columns= ["Febrero 2024"])
#a_feb.to_csv("airbnb_feb_2024")
```

```python
datos_airbnb = pd.concat([a_marzo, a_abril, a_mayo, a_junio, a_julio, a_agosto, a_sept, a_oct, a_nov, a_dic, a_enero, a_feb], axis=1)
#datos_airbnb = datos_airbnb.astype(int)
datos_airbnb = datos_airbnb.fillna(method="ffill")
datos_airbnb
```

Para ver cual es el rango principal entre los que se encuentran los precios los separamos por rangos

```python
rango_50_150 = []
rango_151_250 = []
rango_251_350 = []
rango_351_450 = []
rango_451_600 = []
rango_resto = []


for i in datos_airbnb.columns:
    for dato in range(len(datos_airbnb[i])):
    
        if datos_airbnb[i][dato] <= 150:
            rango_50_150.append(datos_airbnb[i][dato])
        elif datos_airbnb[i][dato] > 150 and datos_airbnb[i][dato] <= 250:
            rango_151_250.append(datos_airbnb[i][dato])
        elif datos_airbnb[i][dato] > 250 and datos_airbnb[i][dato] <= 350:
            rango_251_350.append(datos_airbnb[i][dato])
        elif datos_airbnb[i][dato] > 350 and datos_airbnb[i][dato] <= 450:
            rango_351_450.append(datos_airbnb[i][dato])
        elif datos_airbnb[i][dato] > 450 and datos_airbnb[i][dato] <= 600:
            rango_451_600.append(datos_airbnb[i][dato])
        else:
            rango_resto.append(datos_airbnb[i][dato])      
    
```

```python
rango1 = pd.DataFrame(rango_50_150, columns= ["50 - 150"])
rango2 = pd.DataFrame(rango_151_250, columns= ["151 - 250"])
rango3 = pd.DataFrame(rango_251_350, columns= ["251 - 350"])
rango4 = pd.DataFrame(rango_351_450, columns= ["351 - 450"])
rango5 = pd.DataFrame(rango_451_600, columns= ["451 - 600"])
rango6 = pd.DataFrame(rango_resto, columns= ["resto"])

rangos_airbnb = pd.concat([rango1, rango2, rango3, rango4, rango5, rango6], axis=1)
rangos_airbnb
```

### Creamos data frame de hoteles y ordenamos datos

```python
h_marzo = pd.read_csv("hotel_marzo_2023", index_col= 0)
h_abril = pd.read_csv("hotel_abril_2023", index_col= 0)
h_mayo = pd.read_csv("hotel_mayo_2023", index_col= 0)
h_junio = pd.read_csv("hotel_junio_2023", index_col= 0)
h_julio = pd.read_csv("hotel_julio_2023", index_col= 0)
h_agosto = pd.read_csv("hotel_agosto_2023", index_col= 0)
h_sept = pd.read_csv("hotel_sept_2023", index_col= 0)
h_oct = pd.read_csv("hotel_oct_2023", index_col= 0)
h_nov = pd.read_csv("hotel_nov_2023", index_col= 0)
h_dic = pd.read_csv("hotel_dic_2023", index_col= 0)
h_enero = pd.read_csv("hotel_enero_2024", index_col= 0)
h_feb = pd.read_csv("hotel_feb_2024", index_col= 0)
```

```python
datos_hoteles = pd.concat([h_marzo, h_abril, h_mayo, h_junio, h_julio, h_agosto, h_sept, h_oct, h_nov, h_dic, h_enero, h_feb], axis=1)
```

Como de airbnb tenemos 270 datos y de hoteles tenemos muchos más, para hacer una comparación dentro de unos precios medios que podría pagar un público de clase media filtramos para que nos eneñe los precios que estén por debajo de 1000 y se pueda hacer un abuena estimación de precios y ver cual puede salir más rentable

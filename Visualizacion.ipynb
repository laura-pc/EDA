```python
# Importamos todos los paquetes que vayamos a necesitar
from sklearn.datasets import load_iris
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from plotly.offline import init_notebook_mode, iplot, plot
import plotly as py
init_notebook_mode(connected=True)
import plotly.graph_objs as go

sns.set(color_codes=True)

import warnings
warnings.filterwarnings("ignore")
```

```python
df_airbnb = pd.read_csv("Datos_AIRBNB", index_col=0)
df_hotel = pd.read_csv("Datos_HOTEL", index_col=0)
```

```python
plt.figure(figsize=(20,5))
sns.boxplot(df_airbnb, orient="v");
```

```python
plt.figure(figsize=(20,5))
sns.boxplot(df_hotel, orient="v");
```

```python
airbnb_min = df_airbnb.min()
airbnb_max = df_airbnb.max()
airbnb_mean = df_airbnb.mean()

hotel_min = df_hotel.min()
hotel_max = df_hotel.max()
hotel_mean = df_hotel.mean()
```

```python
trace1 = go.Scatter(
                    x = df_airbnb.columns,
                    y = airbnb_min,
                    name = 'Mínimo Airbnb',
                    mode= 'lines+markers',
                    marker = dict(color = 'rgba(16, 112, 2, 0.8)'),
                    text = "min airbnb")

trace2 = go.Scatter(
                    x = df_hotel.columns,
                    y = hotel_min,
                    name = 'Mínimo Hoteles',
                    mode= 'lines+markers',
                    marker = dict(color = 'rgba(80, 26, 80, 0.8)'),
                    text ="min hotel")

data = [trace1, trace2]

layout = dict(title = 'Precios mínimos de Airbnb vs Hoteles',
             xaxis= dict(ticklen=5, ticks="outside", tickson="boundaries",), 
           )

fig = go.Figure(data = data, layout=layout)

iplot(fig)
```

```python
fig.write_html("prueba.html")
```
